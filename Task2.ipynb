```python
# Citation: Code adapted from https://www.kaggle.com/code/plavak10/ecg-heartbeat-classification
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('ecg_dataset/mitbih_train.csv')
df2 = pd.read_csv('ecg_dataset/mitbih_test.csv')
```

```python
dfs = [df1,df2]
for df in dfs:
    df.columns = list(range(len(df.columns)))
```

```python
data = pd.concat(dfs,axis=0).sample(frac=1.0,random_state=1).reset_index(drop=True)
data
```

```python
data.rename(columns={data.columns[-1]:'Target'},inplace=True)
data
```

```python
X = data.drop('Target',axis=1)
y = data['Target']
```

```python
# Check the number of classes here
list(set(y))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
models = {
    "Decision Tree": DecisionTreeClassifier(max_depth=5),
    "Random Forest": RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42),
    "Logistic Regression": LogisticRegression(multi_class='multinomial', solver='lbfgs'),
    "Neural Network": MLPClassifier(hidden_layer_sizes=(100,), max_iter=300, activation='relu', solver='adam')

}
```

```python
for name,model in models.items():
    model.fit(X_train,y_train)
    print('{} : {}'.format(name,model.score(X_test,y_test)))
```
